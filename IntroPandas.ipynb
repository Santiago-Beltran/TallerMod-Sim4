## Indexación y slicing

Es común querer seleccionar solo ciertas filas o columnas del dataframe: para esto usamos la indexación o el slicing. Por ejemplo, podemos usar números entre corchetes para retornar ciertas filas, y por defecto retorna todas las columnas del dataset.

````python
df[0:3]
````

Podemos elegir filas de columnas en específico usando el nombre de la columna:



````python
df["State"][5:10]
````

&nbsp;

Si queremos elegir múltiples columnas, debemos pasarlas como una lista:



````python
df[["State", "Year", "County", "Uninsured adults"]][19:25]
````

Si usamos el atributo `.iloc` antes de nuestros corchetes, pandas recibirá dos números separados por una coma: el primero para las filas y el segundo para las columnas (así busco por posiciones en vez de por nombres, como si estuvieramos en numpy). Por ejemplo, para seleccionar la fila en el índice 2 y la columna del índice 3:

````python
df.iloc[2,3]
````

````python
df
````

&nbsp;

También podemos usar ":" para elegir múltiples filas o columnas:


````python
df.iloc[:,3] # Todas las filas de la columna con índice 3
````

&nbsp;



````python
df.iloc[0:3,:] # Primeras 3 filas, todas las columnas
````

&nbsp;



````python
df.iloc[120:126,2:8] # Filas 120 a 125 de columnas 2-8 (zero-indexed)
````

La **i** de `iloc` hace referencia a **integer**. 
Si queremos usar los nombres de las columnas, usamos el atributo `.loc`.

````python
df.loc[0:3,"Division"]
````

````python
### Notar que esto es equivalente a lo anterior
df["Division"][:4]
````

````python
## Múltiples columnas
df.loc[0:3,["State","County","Food environment index"]]
````

### Asignación por índexación

Cuando queremos asignar un valor a una celda o región del dataframe, es una buena práctica usar `.loc` o `.iloc` para seleccionar dicha región. Esto asegura que se hagan lo cambios sobre el dataframe y no sobre una copia.

En el siguiente código, la opción que no usa el .loc podría a veces fallar dependiendo de los índices del dataframe, sobre todo cuándo se tienen múltiples (esto ocurre por la forma en que se procesa por debajo en Python lo que está haciendo pandas. Para evitar riesgos se recomienda usar el .loc).

````python
ex = df.copy()

# Es mejor hacer esto
ex.loc[0,"County"] = "New County"

# Lo siguiente a veces puede no funcionar
#ex["County"][0] = "New County"

print(ex.loc[0,"County"])
````

### Filtrado (súper importante y usado)


Una de las tareas más comunes en analítica de datos, en la parte de preparación, es el filtrado de estos. Prácticamente todos los conjuntos de datos que usaremos requieren de ciertos filtros previos antes de estar listos para su análisis.  

Incluso puede haber roles que se dedican a responder preguntas de negocio a través de filtros eficientes y precisos para obtener los mejores resultados.

Para filtrar el dataset basado en una condición lógica, usamos unos corchetes anidados. En el siguiente ejemplo:

* En los corchetes internos, `df["State"]=="RI"` verifica en que partes de la columna State el valor es igual a "RI"
* El corchete externo `df[ ... ]` usa el resultado del interno para filtrar las filas


````python
df[df["State"] == "RI"]
````

También se puede usar múltiples condiciones:

````python
# Esto sería un operador O
df[(df["GEOID"] == 44001) | (df["County"] == "Kent County")]
````

### Estadísticos descriptivos

Los estadísticos descriptivos (promedio, mínimo, máximo, etc.) son esenciales para explorar y entender un conjunto de datos. Cuando los generamos usando pandas, usualmente queremos que nuestros datos se agreguen por un cierto grupo (región, estado, país, etc).

Para lograr esto, se usa el método .groupby(). 

En el ejemplo, encadenamos dos métodos, para primero agrupar por estado y luego obtener el promedio de costos de salud (este tipo de consultas son muy comunes en análisis de datos y en SQL, y muchas veces, a pesar de ser tan simples, pueden ser de gran utilidad y valor para las empresas):

````python
avgStCosts = df.groupby("State")["Health care costs"].mean()
avgStCosts.head()
````

&nbsp;

Si queremos ordenar la lista obtenida de mayor a menor costos, podemos usar  `.sort_values()` con el resultado obtenido. Además, encadenemos .round apra redondear a dos decimales.

````python
avgStCosts = avgStCosts.sort_values(ascending = False).round(2)
avgStCosts.head()
````

&nbsp;

Ordenar los dataframes en pandas usualmente reordenará también los índices de estos. Notar que pasa con los índices luego de ordenarlo:


````python
nuevo_df = df.sort_values(by = "Poor or fair health").head()
````

````python
nuevo_df
````

&nbsp;

Si se quiere volver a generar el índice con el nuevo ordenamiento, se usa el método  `reset_index()`, usualmente con el argumento "drop=True" para que no se me genere una columna nueva llamada Index y para que los índices arranquen en orden desde cero.

#### Ojo:
Tener unos índices que no están en orden sucesivo (de 1 en 1) y ascendente puede traer problemas en ciertas ocasiones, por lo que hay que usar este reseteo de índice en ciertos casos para evitar problemas a futuro. (Por ejemplo, si no resetea los índices, en Pandas al querer ir a la posición 0, puede que traiga en realidad el índice 0 que puede estar en otra posición muy diferente).

````python
### Con la opción drop=True, se elimina el índice anterior completamente y los nuevos índices quedan en orden ascendente
nuevo_df = nuevo_df.sort_values(by = "Poor or fair health").reset_index(drop=True).head()
nuevo_df
````

## "Big Data" y pandas 

Pandas puede también leer csv por "lotes", para así lograr cargar conjuntos de datos muy grandes, que si se leyeran todos al mismo tiempo podrían generar problemas de RAM.

En el siguiente código, usar `chunksize` y `iterator=True` genera un flujo de lotes (chunks) de 1000 filas cada uno del dataset, para realizar la carga a través de estos lotes. En este ejemplo no es necesario hacer carga por lotes, pero si el dataset fuera de 61millones de filas, probablemente si se requieriera cargarlo así (más adelante en el curso veremos un ejemplo de un dataset tan pesado que puede agotar la RAM del computador al cargarlo entero).


````python
# Lista de los lotes
chunk_list = []

# Leer lotes de 1000 filas uno por uno y almacenar solo las filas donde el estado es NC
# La tabla final tendrá mucho menos filas que la tabla de origen porqeu ese están filtrando
# Pero tengo que hacer este filtro así por lotes para no colapsar la RAM si la tabla origen fuera muy grande
for chunk in pd.read_csv("CountyHealthData_2014-2015.csv", chunksize=1000, iterator=True):
    nc_rows = chunk[chunk["State"]=="NC"]
    chunk_list.append(nc_rows)

#Combinar todos los lotes en un solo dataframe que tenga los resultados
nc_df = pd.concat(chunk_list, ignore_index=True)
nc_df.head()
````
